```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure plots are displayed inline
%matplotlib inline

# Load the dataset (make sure the path is correct)
data = pd.read_excel(r'C:\Users\eljo6\Desktop\Project1_Data_Analytics\Data\world_happines.xlsx',)

# Histogram of Happiness Scores (adjust column name if necessary)
plt.figure(figsize=(10, 6))
sns.histplot(data['Happiness score'], bins=20, kde=True)
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()  # Make sure to include this line

# Display basic information about the dataset
print(data.info())

# Display statistical summary of the dataset
print(data.describe())

# Check for missing values
print(data.isnull().sum())

# Scatter plot of Happiness Score vs GDP per Capita
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Explained by: GDP per capita', y='Happiness score')
plt.title('Happiness Score vs GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()

# Box plot of Happiness Scores by Region (if applicable)
plt.figure(figsize=(25, 6))
sns.boxplot(x='Country', y='Happiness score', data=data)
plt.title('Happiness Scores by Region')
plt.xlabel('Region')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.regplot(data=data, x='Explained by: GDP per capita', y='Happiness score')
plt.title('Happiness Score vs GDP per Capita with Trend Line')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()

# Scatter plot of Happiness Score vs Social Support
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Explained by: Social support', y='Happiness score')
plt.title('Happiness Score vs Social Support')
plt.xlabel('Social Support')
plt.ylabel('Happiness Score')
plt.show()

# Calculate the correlation between Happiness Score and Social Support
social_support_correlation = data['Happiness score'].corr(data['Explained by: Social support'])
print(f"Correlation between Happiness Score and Social Support: {social_support_correlation:.2f}")

# Calculate the correlation between Happiness Score and GDP per capita
GDP_correlation = data['Happiness score'].corr(data['Explained by: GDP per capita'])
print(f"Correlation between Happiness Score and GDP per capita: {GDP_correlation:.2f}")

```

## Key Findings and Insights

- **GDP and Happiness**: We observed a strong positive correlation between GDP per Capita and Happiness Score, indicating that higher GDP is **generally** associated with higher happiness levels.
- Our analysis reveals a strong positive correlation (0.76) between **GDP per capita** and **Happiness Score**. This suggests that countries with higher GDP per capita tend to report higher happiness levels.
## Regional Differences
Scandinavean countries have notably higher happiness scores compared to other regions. This may be due to some factors like:
- **Strong Social Welfare Systems**: These countries provide universal healthcare, free education, and other support services that reduce economic stress and contribute to higher well-being.
- **Balanced Work Culture**: Policies promoting work-life balance help maintain high levels of life satisfaction.
- **Low Income Inequality**: With more even income distribution, more people have access to resources, contributing to overall happiness.
- **High Levels of Social Trust**: Trust in society and institutions is high, fostering a sense of security and community.
### Correlation Between Social Support and Happiness

Our analysis reveals a strong positive correlation (0.78) between **Social Support** and **Happiness Score**. This suggests that countries with higher levels of social support tend to report higher happiness levels.

#### Possible Explanations for This Correlation:
- **Emotional Well-being**: Social support networks reduce stress and improve mental health, both of which contribute to happiness.
- **Financial and Practical Assistance**: Support during challenging times provides a safety net, which can reduce worry and enhance well-being.
- **Sense of Belonging**: Being part of a supportive community fulfills a basic human need, increasing life satisfaction.

Overall, this strong correlation underscores the importance of social connections and support in promoting happiness.

### Observations and Limitations

### Outlier Analysis: Venezuela

Venezuela appears as an outlier in our analysis, with a **GDP per Capita of 0.0** but a **Happiness Score of 4.95**. Possible explanations include:

- **Data Inconsistencies**: The reported GDP per Capita of 0.0 may indicate missing or unreliable data due to the country’s economic crisis.
- **Social Resilience**: Despite economic hardship, social factors and resilience may contribute to the country's moderate Happiness Score.

This outlier underscores that Happiness Scores can be influenced by factors beyond just economic measures, such as community support and social ties.

- **Data Limitations**: The data only includes certain countries, and factors like cultural biases in survey responses might influence the scores.
